### Étape 3 : Téléchargement du Dataset Médical
Télécharge les IRM depuis Kaggle et les place automatiquement dans la bonne structure `data/raw/brain_mri/`.

```python
import kagglehub
import shutil
from pathlib import Path

print(" Téléchargement du dataset 'Brain Tumor Classification (MRI)' depuis Kaggle...")
path = kagglehub.dataset_download("sartajbhuvaji/brain-tumor-classification-mri")

# Création des dossiers cibles
dst = Path("data/raw/brain_mri")
dst.mkdir(parents=True, exist_ok=True)

# Copie des données
shutil.copytree(path, dst, dirs_exist_ok=True)

print(f" Dataset prêt dans le répertoire : {dst.absolute()}")
```

### Étape 4 : Configuration du PYTHONPATH
Cette étape est cruciale pour que Colab (et Papermill) reconnaisse nos modules internes situés dans le dossier `src/`.

```python
import sys
import os

project_path = "/content/pinkcc-ct-lesion-segmentation"
src_path = f"{project_path}/src"

# Pour le notebook actuel
if src_path not in sys.path:
    sys.path.insert(0, src_path)

# Pour l'exécution externe via ligne de commande (Papermill)
current_pythonpath = os.environ.get('PYTHONPATH', '')
os.environ['PYTHONPATH'] = f"{current_pythonpath}:{src_path}"

print(" Variables d'environnement configurées.")
```
